```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imigrantes_canada.csv')
```

```python
df.head()
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
top_10
```

```python
sns.barplot(top_10, x='Total', y='País')
```

```python
ax = sns.barplot(top_10, x='Total', y='País')

ax.set(title= 'Países com maior imigração para o Canadá \n 1980 a 2013',
      xlabel= 'Número de imigrantes',
      ylabel = '')


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(top_10, x='Total', y='País')

ax.set_title('Países com maior imigração para o Canadá \n 1980 a 2013',loc= 'left', fontsize=15)

ax.set_xlabel( 'Número de imigrantes', fontsize=14)
ax.set_ylabel('')


plt.show()
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(top_10, x='Total', y='País', palette=palette)
    ax.set_title('Países com maior imigração para o Canadá \n 1980 a 2013',loc= 'left', fontsize=15)
    ax.set_xlabel( 'Número de imigrantes', fontsize=14)
    ax.set_ylabel('')
 
    
    plt.show()
```

```python
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('rocket')
```

```python
gerar_grafico_paleta('coolwarm')
```

```python
gerar_grafico_paleta('hls')
```

```python
sns.set_theme(style='dark')

gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')

gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')

gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')

gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(top_10, x='Total', y='País', palette='tab10')

ax.set_title('Países com maior imigração para o Canadá \n 1980 a 2013', fontsize=15)
ax.set_xlabel( 'Número de imigrantes', fontsize=14)
ax.set_ylabel('')
sns.despine()

plt.show()
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
df.set_index('País', inplace=True)
```

```python
df
```

```python
sns.set_theme()
sns.set_palette('Dark2')

fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.lineplot(df.loc['Brasil', anos], label='Brasil', lw=3)
ax = sns.lineplot(df.loc['Argentina', anos], label='Argentina', lw=3)
ax = sns.lineplot(df.loc['Peru', anos], label='Peru', lw=3)
ax = sns.lineplot(df.loc['Colômbia', anos], label='Colômbia', lw=3)

ax.set_title('Imigração dos maiores países da América do Sul\npara o Canadá de 1980 a 2013', loc='left', fontsize=20)
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

ax.legend(title='Países', loc='upper right', bbox_to_anchor=(1.18, 1.02))

plt.show()
```
