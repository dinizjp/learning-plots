```python
import plotly.express as px
import pandas as pd 
```

```python
df = pd.read_csv('imigrantes_canada.csv')
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))

brasil = df.loc['Brasil', anos]

brasil_dict = { 'ano' : brasil.index.tolist(), 'numero_imigrantes' : brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
fig= px.line(dados_brasil, x='ano', y='numero_imigrantes')

fig.update_layout(
    xaxis={'tickangle': -45}, 
    xaxis_title='Ano',
    yaxis_title='Número de imigrantes'
)

fig.show()
```

```python
fig.show("png",scale=6)
```

```python
fig= px.line(dados_brasil, x='ano', y='numero_imigrantes',
            title='Imigração do Brasil para o Canadá de 1980 a 2013')

fig.update_traces(line_color='black', line_width=2)

fig.update_layout(width=1000, height=500, 
                 xaxis={'tickangle':-45},
                 font_family='Monospace', 
                 font_size=14,
                 font_color='grey',
                 title_font_color='black',
                 title_font_size=22,
                 xaxis_title='Ano',
                 yaxis_title='Número de imigrantes')

fig.show()
```

```python
fig.show("png",scale=5)
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
df_america_sul_clean = america_sul.drop(['Continente', 'Região', 'Total'], axis=1)

america_sul_final = df_america_sul_clean.T
```

```python
america_sul_final
```

```python
fig = px.line(america_sul_final, x=america_sul_final.index, y=america_sul_final.columns, color='País',
             title='Imigração dos países da América do Sul para o Canadá de 1980 a 2013',
             markers=True)

fig.update_layout(
    xaxis={'tickangle': -45}, 
    xaxis_title='Ano',
    yaxis_title='Número de imigrantes'
)

fig.show()
```

```python
fig.show("png",scale=5)
```

```python
fig.write_html('imigracao_america_sul.html')
```
