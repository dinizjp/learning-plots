```python
import pandas as pd
```

```python
df = pd.read_csv('imigrantes_canada.csv')

df.head()
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
df
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = { 'ano' : brasil.index.tolist(), 'numero_imigrantes' : brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))

plt.plot(dados_brasil['ano'], dados_brasil['numero_imigrantes'])

plt.title('Imigração do Brasil para o Canadá')

plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')

plt.xticks(['1980', '1985', '1990', '1995','2000','2005','2010'])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(dados_brasil['ano'], dados_brasil['numero_imigrantes'])

ax.set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))

axs[0].plot(dados_brasil['ano'], dados_brasil['numero_imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(dados_brasil['numero_imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá \n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()


plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(15,7))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos 4 maiores Países da América do Sul para o Canadá \n 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')


axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colombia')


axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')
   
    
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')
    
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    
    
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
    
ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)
    
    
plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul.sort_values('Total', ascending=False)
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
fig, ax = plt.subplots(figsize=(14,5))

ax.bar(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá de 1980 a 2013', loc='left', fontsize=15)
ax.set_xlabel('')
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
fig, ax = plt.subplots(figsize=(14,5))

ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá de 1980 a 2013', loc='left', fontsize=15)
ax.set_xlabel('Número de imigrantes')
ax.set_ylabel('', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    
    else:
        cores.append('silver')



fig, ax = plt.subplots(figsize=(14,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes \n para o Canadá de 1980 a 2013', loc='left', fontsize=15)
# ax.set_xlabel('Número de imigrantes')
ax.set_ylabel('', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v, in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color ='black', fontsize=10, ha='left', va='center' )
    
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```
